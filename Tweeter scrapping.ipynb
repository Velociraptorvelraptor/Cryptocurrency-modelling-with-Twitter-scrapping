```python
import pandas as pd
import urllib.request
import datetime
try:
    import GetOldTweets3 as got
    print("GetOldTweets3 imported succesfully")
except(KeyError):
    print("Sorry, unable to download")
```

```python
tweets=[]
dates=[]
username=[]
users=['iamjosephyoung',
'Bitcoin',
'BitcoinMagazine',
'btc',
'BTCNewsletter',
'BTCTN',
'coindesk',
'crypto',
'MoonOverlord',
'TuurDemeester',
'CryptoCoinsNews',
'LandM_Marius',
'kerooke',
'loomdart',
'GemCrypto',
'coinnounce',
'MatiGreenspan',
'JWilliamsFstmed']

n_tweets=30

for i in users:
    try:
        print("Scrapping for %s..."%i)
        starting_time = datetime.datetime.now()
        tweetCriteria = got.manager.TweetCriteria().setUsername(i)\
                                           .setTopTweets(True)\
                                           .setMaxTweets(n_tweets)
       
        for j in range(n_tweets):
            tweet = got.manager.TweetManager.getTweets(tweetCriteria)[j]
            tweets.append(tweet.text)
            dates.append(tweet.date)
            username.append(tweet.username)
        finishing_time = datetime.datetime.now()
        delta = finishing_time - starting_time
        print("\t\t\t\t...and done, took %s seconds."%delta )
    except:
        print("Wrong user name given")

    
```

```python
df_tweets=pd.DataFrame(list(zip(dates, tweets, username)), columns=['dates', 'tweets', 'username'])
df_tweets.sort_values(by='dates', ascending=False)
```

```python
df_tweets.to_csv('data', index=False)
```
